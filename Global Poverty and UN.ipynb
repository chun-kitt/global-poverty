<a id='1.1'><h3>1.1 Background and Overview </h3></a>

Even though global poverty rates have been cut by more than half since 2000, countless people are still trapped in extreme poverty, as evidenced by the latest figures which puts the share of world population in this state at 10% in 2015 (UN SDG). It is understood that short-term and occasional donations are not sustainable and long-term development plans are essential to successfully eradicate poverty. With UN being one of the largest inter-governmental organisations, we would like to explore how funds are allocated in recent years and whether they have improved the poverty rate in these poverty-stricken countries over the past 3 decades.

UN Development Programme is one of the programmes that the UN has to help participating countries to improve their standards of living. In particular, we focus on one of the sustainable development goals set by the UN, which is: “To eradicate extreme poverty by the year 2030”.


To evaluate the changes in standard of living, we will be using changes in poverty rate across countries (quantitative indicator) and HDI (qualitative indicator). Our insights will give us clues to how efficient and effective UN and countries have been in eradicating poverty over the years. We believe that global poverty concerns everybody as the world aims to solve the unequal wealth distribution, therefore every one of us should be a key stakeholder to this. Nonetheless, the UN, large non-profit organisations, governments and economists would be more interested in this aspect.

#### Calculate Total Aid Disbursement Received Per Capita (1990-2013)

```python
df_fund.drop(['Series Code', 'Country Code','2014 [YR2014]','2015 [YR2015]','2016 [YR2016]','2017 [YR2017]',
              '2018 [YR2018]'], axis = 1 , inplace = True)
df_fund['Total' + ' ' + 'ODA' + ' ' + 'Per' + ' ' + 'Capita' + " " + "(1990-2013)"] = df_fund.sum(axis=1)
df_fund = df_fund.iloc[:5]

# Reindex 
df_fund = df_fund.reindex([0,1,2,4,3])
df_fund.reset_index(inplace=True)

df_fund.head()
```

<a id='3.1.2'><h3> 3.1.2 Poverty Rates (1990-2013) </h3></a>


Poverty rate data is evaluated based on proportion of population living under $1.90 per day

####  Importing and Cleaning Data

```python
df_poverty = pd.read_csv('Poverty Rates.csv')
df_poverty = df_poverty.drop([1,6]).reset_index(drop=True)   # Remove Europe and Central Asia 
df_poverty = df_poverty.drop(['2015'],axis=1).reset_index(drop=True)
df_poverty.head()
df_poverty.replace('..',value= 0,inplace=True)


df_poverty.head()

```

#### Calculate the Change in Poverty Rate from 1990 to 2013

```python

df_poverty['Overall' + ' ' + 'Decrease ' +'' + 'Rate' + ' ' + '(%)'] = abs((df_poverty['2013'].astype(float) - 
                                           df_poverty['1990'].astype(float))/df_poverty['1990'].astype(float))*100 

df_poverty.head(10)
```

<a id='3.2'><h3> 3.2 Analysis and Evaluation </h3></a> 
<a id='3.2.1'><h3> 3.2.1 Analysing using Fund Efficiency</h3></a> 

#### We analyse how efficient fund has been in decreasing poverty rate in each region
- We first calculated the "Percentage Decrease in Poverty Per Dollar Disbursed Per Capita" in each region from 1990-2013. This is done by using <b> Formula 1 </b> . 

$$
\text{Formula 1} = \frac{\text{Overall % Decrease in Poverty Rate}_{1990-2013}}{\text{Total ODA Per Capita}_{1990-2013}}
$$

- We then normalise this value for each region in the range of 0 to 1 and termed this as "Fund Efficiency". This is done by using <b> Formula 2 </b>. 

$$
\text{Formula 2} = \frac{\text{% Decrease in Poverty Per Dollar Per Capita - % Decrease in Poverty Per Dollar Per Capita}_{Min}}{\text{% Decrease in Poverty Per Dollar Per Capita}_{Max} - \text{% Decrease in Poverty Per Dollar Per Capita}_{Min}}
$$

```python
# Percentage Decrease in Poverty Per Dollar Disbursed Per Capita

df_poverty['PCTdecrease_PerDollarPerCapita'] = df_poverty['Overall Decrease Rate (%)']/ df_fund['Total ODA Per Capita (1990-2013)']


# Fund Efficiency for Poverty 
df_poverty['Fund_Efficiency'] = (df_poverty['PCTdecrease_PerDollarPerCapita'] - df_poverty['PCTdecrease_PerDollarPerCapita'].min()) / (df_poverty['PCTdecrease_PerDollarPerCapita'].max() -
                                                                                                                                     df_poverty['PCTdecrease_PerDollarPerCapita'].min())


# Filtered data

df_poverty_filtered = df_poverty[['Region', '1990', '2013', 'Overall Decrease Rate (%)',
                                  'PCTdecrease_PerDollarPerCapita', 'Fund_Efficiency']]


df_poverty_filtered.head()
```

<a id='4.1'><h3> 4.1. Data Cleaning and Output</h3></a> 
<a id='4.1.1'><h3> 4.1.1 HDI Dataset (1990-2013) </h3></a> 

#### Importing HDI Dataset 

```python
HDI_df = pd.read_csv('HDI.csv', encoding='latin-1')
HDI_df.head()
HDI_df = HDI_df.fillna(0)
```

#### Find the total HDI for Each Region 

```python
# South Asia #
HDI_SouthAsia_df = HDI_df[HDI_df['Country'].isin(SAsia)]
HDI_SouthAsia_total = HDI_SouthAsia_df.sum().reset_index(name='SouthAsia Total')
 
# Latin America and Caribbean Countries #
HDI_LatinAmerica_df = HDI_df[HDI_df['Country'].isin(LatinA)]
HDI_LatinAmerica_total= HDI_LatinAmerica_df.sum().reset_index(name='LatinAmerica Total')

# Middle East Asia and North Africa #    
HDI_MiddleEastAsia_NAfrica_df = HDI_df[HDI_df['Country'].isin(ME_NA)]
HDI_MiddleEastAsia_NAfrica_total= HDI_MiddleEastAsia_NAfrica_df.sum().reset_index(name='MiddleEastAsia_NAfrica Total')

# East Asia and Pacific # 
HDI_EastAsiaPacific_df = HDI_df[HDI_df['Country'].isin(EAsia_P)]
HDI_EastAsiaPacific_total= HDI_EastAsiaPacific_df.sum().reset_index(name='EastAsiaPacific Total')
 
# Sub Saharan Africa # 
HDI_SubSaharanAfrica_df = HDI_df[HDI_df['Country'].isin(Sub_SA)]
HDI_SubSaharanAfrica_total= HDI_SubSaharanAfrica_df.sum().reset_index(name='SubSaharanAfrica Total')
 
# Combine DF #
Combine1_df = HDI_SubSaharanAfrica_total.join(HDI_MiddleEastAsia_NAfrica_total,
                           lsuffix = 'HDI_SubSaharanAfrica_total', rsuffix = 'HDI_MiddleEastAsia_NAfrica_total')
Combine2_df = Combine1_df.join(HDI_SouthAsia_total, lsuffix = 'Combine1_df', rsuffix = 'HDI_SouthAsia_total')
Combine3_df = Combine2_df.join(HDI_LatinAmerica_total, lsuffix = 'Combine2_df', rsuffix = 'HDI_LatinAmerica_total')
HDI_Regions_Total= Combine3_df.join(HDI_EastAsiaPacific_total, lsuffix = 'Combine3_df', rsuffix = 'HDI_EastAsiaPacific_total')

HDI_Regions_Total = HDI_Regions_Total.iloc[2:]
HDI_Regions_Total.drop(columns=['indexHDI_MiddleEastAsia_NAfrica_total','indexCombine2_df','indexHDI_LatinAmerica_total','index'], inplace=True)
HDI_Regions_Total.rename(columns={"indexHDI_SubSaharanAfrica_total": "Year"}, inplace=True)
HDI_Regions_Total = HDI_Regions_Total.transpose()
HDI_Regions_Total = HDI_Regions_Total.rename(columns=HDI_Regions_Total.iloc[0]).drop(HDI_Regions_Total.index[0])
HDI_Regions_Total.drop(columns=['2014','2015','2016','2017'], inplace = True)
HDI_Regions_Total = HDI_Regions_Total.reindex(['EastAsiaPacific Total','LatinAmerica Total', 'MiddleEastAsia_NAfrica Total',
                          'SouthAsia Total', 'SubSaharanAfrica Total'])


HDI_Regions_Total.head()
```

<a id='4.2'><h3> 4.2 Analysis and Evaluation </h3></a> 
<a id='4.2.1'><h3> 4.2.1 Analysing using 'HDI Fund Efficiency' </h3></a> 

#### We analyse how efficient fund has been used in decreasing poverty rate in each region
- We first calculated the "Percentage Increase in HDI Per Dollar Disbursed Per Capita" in each region from 1990-2013. This is done by using <b> Formula 3 </b>.  

$$
\text{Formula 3} = \frac{\text{Overall % Increase in HDI}_{1990-2013}}{\text{Total ODA Per Capita}_{1990-2013}}
$$

- We then normalise this value for each region in the range of 0 to 1 and termed this as "HDI Fund Efficiency". This is done by using <b> Formula 4 </b>.

$$
\text{Formula 4} = \frac{\text{% Increase in HDI Per Dollar Per Capita - % Increase in HDI Per Dollar Per Capita}_{Min}}{\text{% Increase in Poverty Per Dollar Per Capita}_{Max} - \text{% Increase in Poverty Per Dollar Per Capita}_{Min}}
$$

```python
# Overall % Increase in HDI 

HDI_Regions_Total['Overall_Increase_Rate (%)'] = (HDI_Regions_Total['2013'] - HDI_Regions_Total['1990']) / (HDI_Regions_Total['1990']) *100

HDI_Regions_Total.reset_index(inplace=True)


# % Increase in HDI Per Dollar Disbursed Per Dollar 
HDI_Regions_Total['PCTIncrease_PerDollarPerCapita'] = (HDI_Regions_Total['Overall_Increase_Rate (%)'].astype(int))/(df_fund['Total ODA Per Capita (1990-2013)'].astype(int))


# HDI_Fund Efficiency 
HDI_Regions_Total['HDI_FundEfficiency'] = (HDI_Regions_Total['PCTIncrease_PerDollarPerCapita'] - HDI_Regions_Total['PCTIncrease_PerDollarPerCapita'].min()) / (HDI_Regions_Total['PCTIncrease_PerDollarPerCapita'].max() -
                                                                                                                                     HDI_Regions_Total['PCTIncrease_PerDollarPerCapita'].min())


# Filter other years away
HDI_Regions_Total_Filtered = HDI_Regions_Total[['index', '1990', '2013', 'Overall_Increase_Rate (%)',
                             'PCTIncrease_PerDollarPerCapita', 'HDI_FundEfficiency']]


HDI_Regions_Total_Filtered = HDI_Regions_Total_Filtered.rename(columns= {'index':'Region'})

HDI_Regions_Total_Filtered.head()
```

#### <u> Overall Evaluation </u>

There doesn't seem to be a clear pattern on how UN is disbursing fund in the recent years. It is neither disbursing a higher amount per capita to region which has a lower "GDP Per Capita" nor regions which have higher "Fund Efficiency". The results show us that Latin America & Caribbean and Middle East & North Africa is receiving the most "Funds Per Capita" from the UN in recent years despite having a historically lower "Fund Efficiency" based on Aid Disbursement.

Nonetheless, this doesn't mean that regions that are less efficient should be receiving lesser funds. However, there is a need to resolve the underlying factors which are resulting in a lower fund efficiency. This seems to be the root cause of problem for most countries that are trapped in poverty. This is shown by the bar chart whereby regions which have the lower "Increase in Poverty Rate" or lower "Increase in HDI" are also those which have relatively low fund efficiency.

If these problems are not resolved, then a higher allocation of fund assistance towards these less efficient regions will be ineffective in resolving poverty. Hence, as the UN Disburses more funds to these region, it should also considered how funds could be used to resolve the problems which are resulting in lower efficiency usage of fund. Such low efficiency could be resulted from various factors such as poor institutions whereby a high corruption rate resulted in these funds not being channelled effectively to improve the interest of general public in the country. It could also be due to short-sighted objectives whereby funds are largely allocated to resolve short-term issues and does not address poverty in the long run, resulting in a low efficiency of fund usage.     

In conclusion, we believe that there is more to be done than disbursing more funds to alleviate poverty. Our prediction shows that UN might not be able to achieve its objective of "Eradicating Extreme World Poverty by 2030" because Sub Saharan Africa has a significantly high postive poverty rate (based on international poverty line). The UN and other organisations should also consider how plans could be adopted to improve fund effiency especially in regions such as Sub Saharan Africa, Latin America and Middle East & North Africa which has a significantly and relatively lower effiency of fund usage. On the other hand, South Asia has a relatively higher fund efficiency but it might not be able to eradicate poverty (based on international line) totally by 2030. Hence, more funds could be allocated to it without much consideration to the improvement in fund efficiency as compared to the other three regions.
